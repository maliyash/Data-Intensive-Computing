Problem 1: Define two synthetic vectors of data representing sales over 12 months for 2 items. 
           Compare the two using lines graphs.Run the code multiple time and see randomness of the second set of                          data..)

```r
sales1<-c(12,14,16,29,30,45,19,20,16,19,34,20)
sales2<-rpois(12,20)

print(sales2)

par(bg="cornsilk")

plot(sales1,col="blue",type="o",ylim=c(0,300),xlab="Month",ylab="Sales")
title(main="Sales by Month")

lines(sales2,type="o",pch=22,lty=2,col="red")
grid(nx=NA,ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

Discussion:

The function c() is used to create a vector with values explicitly specified by us.
par(bg="cornsilk") sets the color of plot background as cornsilk.
Sales 2 data is a vector with values assigned according to poison distribution.
Poisson Distribution is a discrete frequency distribution which gives the probability of a number of independent events 
occurring in a fixed time. The key difference between poisson distribution and normal distribution is that the poisson distribution is discrete whereas normal distribution is continuous and When the mean of a Poisson distribution is large, it becomes similar to a normal distribution. so, after changing lambda in rpois(), it is observed that it generates a list of random numbers that have the mean equal to lambda (specified in the rpois() function).
